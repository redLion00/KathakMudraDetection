```python
import os
import ffmpeg
```

```python
ORGINAL_PATH = r'C:\Users\Vivek\projects\KathakMudraDataset'
PATH = r'data'
FFMPEG_PATH = r"C:\ffmpeg\bin\ffmpeg.exe"
```

```python
def count_videos(path):
    total_count = 0
    for files in os.listdir(path):
        count = len(os.listdir(os.path.join(path, files)))
        print(f"{files}: {count} videos")
        total_count += count
    print(f"Total videos: {total_count}")
```

```python
count_videos(ORGINAL_PATH)
```

```python
def convert_to_mp4(ORIGINAL_PATH):

    output_dir = os.path.join(PATH, "Videos")
    os.makedirs(output_dir, exist_ok=True)
    for folder in os.listdir(ORIGINAL_PATH):
        folder_path = os.path.join(ORIGINAL_PATH, folder)
        if not os.path.isdir(folder_path):
            continue

        output_folder = os.path.join(output_dir, folder)
        os.makedirs(output_folder, exist_ok=True)
        for file in os.listdir(folder_path):
            
            input_path = os.path.join(folder_path, file)
            filename, ext = os.path.splitext(file)
            ext = ext.lower()
            if ext not in [".mov", ".mp4"]:
                continue
            output_path = os.path.join(output_folder,filename + ".mp4")
            print(f"Processing: {input_path}")

            (
                ffmpeg
                .input(input_path)
                .output(
                    output_path,
                    vf="scale=1080:1920",
                    vcodec="libx264",
                    acodec="aac"
                )
                .run(
                    overwrite_output=True,
                    cmd=FFMPEG_PATH
                )
            )

            print(f"Saved: {output_path}")

    print("Dataset conversion completed.")
```

```python
convert_to_mp4(ORGINAL_PATH)
```

```python
count_videos(PATH)
```
